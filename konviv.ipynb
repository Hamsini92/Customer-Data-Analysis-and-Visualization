```python
import sqlite3
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
conn = sqlite3.connect('ass.db')
#creation of cursors to execute SQL queries
cur = conn.cursor()

#execute the query for fetching all rows from the customer table
df = pd.read_sql_query("select * from sample",conn)
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
#print(df)
df.dropna(how='all')
print(df)
category_data = df["Category"]
df['Amount'] = df['Amount'].convert_objects(convert_numeric=True)
amount_data = df["Amount"]

print(amount_data)

fig = plt.figure(figsize=[10, 10])
ax = fig.add_subplot(111)
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"]
#explode = (0.1, 0, 0, 0, 0)
pie_wedge_collection = ax.pie(amount_data, labels=category_data,colors=colors,
autopct='%1.1f%%', shadow=True, startangle=400)

for pie_wedge in pie_wedge_collection[0]:
    pie_wedge.set_edgecolor('white')

plt.title("Amount spent for various categories")
plt.show()
get_data()
```
